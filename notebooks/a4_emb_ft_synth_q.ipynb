```python
import json
import re
from tqdm import tqdm
```

```python
DATA_PATH = "../data/processed/proc_docx.jsonl"
```

```python
docs = []
with open(DATA_PATH, "r", encoding="utf-8") as f:
    for line in f:
        data = json.loads(line)
        text = data.get("text", "")
        meta = {
            "doc_id": data.get("doc_id", ""),
            "file_path": data.get("file_path", "")
        }
        if text.strip():
            docs.append({"text": text, "metadata": meta})
```

```python
file_names = []
for doc in docs:
    name = doc['metadata']['file_path'].replace('./data/raw/', '').replace('.docx', '')
    file_names.append(name)
file_names
```

```python
# text = "[Title] Отчет по акции: Кэшбэк — Бакалея, Сибирь. Период акции: 18.12.2025 – 28.7.2025 Регион: Сибирь Категория: Бакалея Механика акции: Кэшбэк "
# match = re.search(r'\[Title\][^.]*\.\s*(.*)', text)
# if match:
#     result = match.group(1)
#     print(result)

# pattern = r'\[Title\][^.]*\.\s*(.*)'
pattern = r'\[Title\][^.]*\.\s*([^.]*)\.'
# pattern = r'\[Title\][^.]*\.\s*([^.]*?(?:\.\d+)*?(?:\s[^.]*)*?)\.(?!\d)'
def get_title(text):
    match = re.search(pattern, text)
    if match:
        result = match.group(1)
        return result
    return ''
```

```python
titles = []
for doc in docs:
    # print(doc['text'])
    t = get_title(doc['text'])
    if 'Период акции' not in t and '[Title]' not in t and '**' not in t:
        # print('Период акции' in t)
        # print(t)
        # print('-'*80)
        titles.append(t)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "Qwen/Qwen2.5-7B-Instruct"

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_name)

prompt = "Give me a short introduction to large language model."
messages = [
    {"role": "system", "content": "You are Qwen, created by Alibaba Cloud. You are a helpful assistant."},
    {"role": "user", "content": prompt}
]
text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
)
model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=512
)
generated_ids = [
    output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
]

response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
```

```python
USER_PROMPT = '''Сгенерируй возможный поисковой запрос, который пользователь мог бы ввести, чтобы найти документ по заголовку.
Вот пример:
заголовок: 'Скидка — Москва — Редкие посетители. апрель 2025'
поисковой запрос: Маркетинговые кампании по региону Москва за последний год по сегменту редких клиентов

Напиши поисковой запрос для заголовка:'''
def build_prompt(title):
    return [
        {"role": "system", "content": "Ты — ассистент, который генерирует поисковые запросы по заголовку документа по теме CVM."},
        {"role": "user", "content": f"{USER_PROMPT} «{title}»"}
    ]
```

```python
print(build_prompt(file_names[0])[1]['content'])
# build_prompt(file_names[0])[1]['content']
```

```python
def get_response(title):
    text = tokenizer.apply_chat_template(
        build_prompt(title),
        tokenize=False,
        add_generation_prompt=True
    )
    model_inputs = tokenizer([text], return_tensors="pt").to(model.device)
    generated_ids = model.generate(
        **model_inputs,
        max_new_tokens=512
    )
    generated_ids = [
        output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
    ]
    response = tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
    
    return response
```

```python
res_lst = []
for p in tqdm(file_names+titles):
    res = get_response(p)
    res_lst.append(res)
```

```python
lst_comb = []
comb = file_names+titles
for i in range(len(comb)):
    lst_comb.append((res_lst[i], comb[i]))
lst_comb
```

```python
def has_chinese(text):
    """Проверяет, есть ли китайские символы в тексте."""
    return bool(re.search(r'[\u4e00-\u9fff]', text))

final_lst = []
# final_dct = defaultdict(list)
for j in lst_comb:
    i, k = j
    if not has_chinese(i):
        if "Поисковой запрос" in i:
            # final_lst.append((k , i.replace('Поисковой запрос: ', '')))
            # final_dct[k].append(i.replace('Поисковой запрос: ', ''))
            final_lst.append({'title': k, 'queries': i.replace('Поисковой запрос: ', '')})
        else:
            # final_lst.append((k, i))
            # final_dct[k].append(i)
            final_lst.append({'title': k, 'queries': i})
```
